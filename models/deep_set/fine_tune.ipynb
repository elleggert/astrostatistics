```python

import torch
import os
```

```python
area = 'north'
gal = 'qso'

best_val = -1000
for model in os.listdir(f"trained_models/{area}/{gal}"):
    try:
        int(model[:-3])
        continue
    except:

        val = float(model[:-3])
        print(val)
        if val > best_val:
            best_val = val

print(best_val)
```

```python
model = torch.load(f"trained_models/{area}/{gal}/{best_val}.pt",
                   map_location=torch.device('cpu'))
```

```python
print(model.state_dict())
```

```python
model.train()
```
